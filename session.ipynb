```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
house_price = pd.read_csv ("F:\\Drive D\\AI Training\\python_venv\\Session\\Dataset\\Housing.csv")
```

```python
house_price.head()
```

```python
def map_yes_or_no(df, columns):
    for col in columns:
        df[col]=df[col].map({'yes':1, 'no':0})
    return df
columns_to_map = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

house_price = map_yes_or_no(house_price, columns_to_map)
```

```python
house_price = pd.get_dummies(house_price,columns=["furnishingstatus"])
```
